## Technical View
The canonical current path is the config-driven pipeline:
- dataset config: `configs/data/datasets.yaml`
- train configs: `configs/train/*.yaml`
- training entry point: `src/train.py`
- evaluation entry point: `src/evaluate.py`
- canonical data module: `src/data/datasets.py`

The historical baseline path remains available through `src/legacy/` for reproducibility and comparison.

```python
from pathlib import Path

def find_repo_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / 'pyproject.toml').exists() and (candidate / 'src').exists():
            return candidate
    raise FileNotFoundError('Could not locate the TyreVisionX repo root from the current working directory.')

repo_root = find_repo_root()
print('Repo root:', repo_root)
key_paths = [
    'README.md',
    'docs/PROJECT_STATUS.md',
    'docs/ARCHITECTURE.md',
    'configs/data/datasets.yaml',
    'configs/train/train_resnet18.yaml',
    'configs/train/train_resnet34.yaml',
    'src/train.py',
    'src/evaluate.py',
    'src/legacy/train_baseline.py',
    'reports/project_status/current_status.md',
]

for rel in key_paths:
    path = repo_root / rel
    print(f'{rel:45} ->', 'present' if path.exists() else 'missing')
```

## Why The Canonical Pipeline Matters

## Future Work Sections
### Anomaly Detection
Simple explanation: instead of only classifying known labels, anomaly detection tries to spot images that look unusual.

Technical detail: this could become useful if defect types are diverse, rare, or poorly labeled.

### Localization
Simple explanation: show where the defect is, not just whether a defect exists.

Technical detail: this is important for operator trust, review workflows, and later segmentation work.

### Segmentation
Simple explanation: mark the exact defect pixels.

Technical detail: segmentation would be a stronger spatial supervision task than image-level classification.

### Multi-view 3D
Simple explanation: combine several views of the same tyre.

Technical detail: this could support surface-aware reasoning and defect projection in future inspection cells.
