## Research Goal

TyreVisionX studies tyre inspection using computer vision. The current implemented task is supervised binary classification: good tyre image versus defective tyre image. The new research direction is good-only anomaly detection, where the model learns normal tyre appearance from good images and flags unusual images for review.

## Main Tracks

1. Supervised baseline: train/evaluate classifiers using good and defect labels.
2. Anomaly baseline: planned good-only model for normality scoring.
3. Web-data curation: planned workflow to collect and triage likely-good tyre images.
4. Future localization: planned defect localization or segmentation once image-level baselines are stable.
