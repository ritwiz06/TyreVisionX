# Data Manifest Audit

Purpose: inspect TyreVisionX manifest files and dataset counts without training a model.

Inputs: CSV manifests under `data/manifests/` and `data/processed/`.

Outputs: tables of manifest availability, schema, label counts, and split counts.

Current status: this notebook is safe to run locally and does not require GPU.

```python
from pathlib import Path
import pandas as pd

manifest_paths = sorted(Path("data").glob("**/*manifest*.csv"))
manifest_paths
```

```python
for path in manifest_paths:
    print("\n---", path)
    try:
        df = pd.read_csv(path)
    except Exception as exc:
        print("Could not read:", exc)
        continue
    print("rows:", len(df))
    print("columns:", list(df.columns))
    if "label_str" in df.columns:
        print("label_str counts:")
        print(df["label_str"].value_counts(dropna=False))
    if "split" in df.columns and "label_str" in df.columns:
        print("split x label:")
        print(pd.crosstab(df["split"], df["label_str"]))
```
