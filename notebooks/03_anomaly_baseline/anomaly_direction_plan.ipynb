## What Anomaly Detection Means Here

Instead of training on both good and defect examples, the anomaly model trains only on good tyre images. It learns what normal tyres look like. At evaluation time, images that look far away from normal get higher anomaly scores.

## Why Not Build the Full Model Yet

The repository first needs clear data contracts, good-only manifest definitions, and a threshold policy. Building the model before those decisions risks producing results that are hard to interpret or reproduce.

## Planned First Baseline

A simple first baseline should use a frozen CNN feature extractor and distance-based scoring. This is easier to explain and debug than a complex deep anomaly model.
