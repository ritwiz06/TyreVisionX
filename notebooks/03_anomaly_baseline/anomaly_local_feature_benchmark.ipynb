Purpose: compare the current `resnet50_knn` anomaly reference against validation-only threshold refinement and ResNet50 local-feature variants.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

comparison_path = Path('reports/anomaly/local_feature_model_comparison.csv')
if not comparison_path.exists():
    raise FileNotFoundError('Run the local-feature benchmark first.')
df = pd.read_csv(comparison_path)
df
```

## False Negative Comparison

The main question is whether local evidence reduces missed defective tyres.

```python
cols = ['variant', 'local_mode', 'test_recall', 'test_precision', 'test_false_negatives', 'test_false_positives']
display(df[cols].sort_values(['test_false_negatives', 'test_false_positives']))

ax = df.plot.bar(x='variant', y=['test_false_negatives', 'test_false_positives'], figsize=(10, 4))
ax.set_ylabel('Count')
ax.set_title('False negatives and false positives by variant')
plt.xticks(rotation=35, ha='right')
plt.tight_layout()
```

## Interpretation Notes

- Threshold refinement may reduce false negatives by accepting more false positives.
- Multi-crop and patch-grid methods test whether small regional evidence is stronger than whole-image pooling.
- The recommended next step should not be based on test tuning; use this as an honest comparison under fixed validation-selected thresholds.
