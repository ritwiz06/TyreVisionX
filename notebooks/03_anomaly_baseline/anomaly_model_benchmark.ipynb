# Anomaly Model Benchmark

Purpose: summarize the controlled D1 anomaly benchmark for TyreVisionX.

Inputs: `reports/anomaly/model_comparison.csv` and artifacts under `artifacts/anomaly/benchmark/`.

Current status: benchmark executed for locally runnable ResNet18 and ResNet50 variants. EfficientNet, ConvNeXt, and ViT are pending local weights and extractor support.

Important rule: thresholds are selected on validation only; test metrics are reported once per variant.

```python
from pathlib import Path
import pandas as pd

comparison_path = Path('../../reports/anomaly/model_comparison.csv')
if not comparison_path.exists():
    comparison_path = Path('reports/anomaly/model_comparison.csv')
df = pd.read_csv(comparison_path)
df
```

```python
cols = ['variant', 'backbone', 'scorer', 'embedding_type', 'test_recall', 'test_precision', 'test_normal_fpr', 'test_false_negatives', 'test_false_positives', 'test_auroc', 'test_auprc']
df[cols].sort_values(['test_false_negatives', 'test_auprc'], ascending=[True, False])
```

## Interpretation

`resnet50_knn` is the current best executed variant. It reduced false negatives from 93 in the original ResNet18 + Mahalanobis baseline to 23.

The simple ResNet18 patch-grid variant did not help enough. This suggests that local-feature work should continue, but the next version should probably use ResNet50 features or a more careful patch strategy.
