# False Negative Overlap Review

Purpose: compare which defective tyres are missed by the `resnet50_knn` reference and by the best local-feature candidate.

Inputs: CSV files produced by `scripts/anomaly/compare_false_negative_sets.py`.

Outputs: fixed cases, still-missed cases, new misses, and an optional contact sheet.

```python
from pathlib import Path
import pandas as pd

overlap_dir = Path('reports/anomaly/false_negative_overlap')
for name in ['false_negatives_fixed_by_candidate.csv', 'false_negatives_still_missed_by_both.csv', 'false_negatives_new_in_candidate.csv']:
    path = overlap_dir / name
    print(name, 'exists=', path.exists())
    if path.exists():
        display(pd.read_csv(path).head(20))
```

## Review Questions

- Are fixed cases visibly local, subtle, or edge-of-image defects?
- Are still-missed cases low contrast or label-ambiguous?
- Did the local-feature candidate introduce new misses that would be unacceptable for recall-critical tyre inspection?
