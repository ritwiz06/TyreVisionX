# High-Recall Error Pattern Review

Purpose: review the false negatives and false positives from the current high-recall anomaly reference, `resnet50_knn_threshold_sweep`.

Inputs:
- `reports/anomaly/high_recall_false_negative_review_table.csv`
- `reports/anomaly/high_recall_false_positive_review_table.csv`

Current status: review assets exist. The qualitative interpretation is cautious because contact sheets are not full-resolution inspection evidence.

```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display, Markdown

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
elif ROOT.name == '03_anomaly_baseline':
    ROOT = ROOT.parent.parent

fn_csv = ROOT / 'reports/anomaly/high_recall_false_negative_review_table.csv'
fp_csv = ROOT / 'reports/anomaly/high_recall_false_positive_review_table.csv'
fn_sheet = ROOT / 'reports/anomaly/high_recall_false_negative_contact_sheet.png'
fp_sheet = ROOT / 'reports/anomaly/high_recall_false_positive_contact_sheet.png'

fn = pd.read_csv(fn_csv)
fp = pd.read_csv(fp_csv)
print('False negatives:', len(fn))
print('False positives:', len(fp))
display(fn[['image_path', 'anomaly_score', 'threshold']])
```

## False Negatives

These are defective images scored below the selected threshold. They are the most important review set because false negatives are safety-critical.

```python
if fn_sheet.exists():
    display(Image(filename=str(fn_sheet)))
else:
    display(Markdown('False-negative contact sheet is missing.'))
```

## False Positives

These are good-labelled images scored above the recall-priority threshold. Many appear to contain strong texture, lighting variation, shadows, sidewall markings, or unusual close-up structure.

```python
if fp_sheet.exists():
    display(Image(filename=str(fp_sheet)))
else:
    display(Markdown('False-positive contact sheet is missing.'))
display(fp[['image_path', 'anomaly_score', 'threshold']].head(20))
```

## Interpretation

- The false negatives appear subtle and close to threshold, so threshold refinement helped but did not fully solve recall.
- The false positives suggest that normal tyre texture, dirt, glare, and strong grooves can look anomalous to the current embedding model.
- This justifies local-feature investigation, but the current lower/mid-level patch-memory benchmark did not improve results.
- Next useful work should inspect patch-score distributions and nearest-neighbor patch examples.
