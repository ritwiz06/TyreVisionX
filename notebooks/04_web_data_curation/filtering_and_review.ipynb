# Filtering And Review Workflow

Purpose: inspect outputs from web-candidate filtering and prepare a review queue.

Inputs: `data/interim/web_curated_candidates/candidates_kept.csv`, `candidates_rejected.csv`, and `candidates_review_needed.csv` after running `filter_candidates.py`.

Outputs: visual counts and review-ready tables.

Current status: runnable when candidate metadata exists; no real web run is included.

```python
from pathlib import Path
import pandas as pd

OUT_DIR = Path("data/interim/web_curated_candidates")
paths = {
    "kept": OUT_DIR / "candidates_kept.csv",
    "rejected": OUT_DIR / "candidates_rejected.csv",
    "review_needed": OUT_DIR / "candidates_review_needed.csv",
}
frames = {}
for name, path in paths.items():
    if path.exists():
        frames[name] = pd.read_csv(path)
    else:
        print(f"Missing {name}: {path} (run filter_candidates.py after candidate collection/download)")
frames.keys()
```

```python
if frames:
    counts = pd.Series({name: len(df) for name, df in frames.items()}).rename("count")
    display(counts.to_frame())
    counts.plot(kind="bar", title="Candidate filtering buckets")
```

```python
if "review_needed" in frames and len(frames["review_needed"]):
    display(frames["review_needed"][["candidate_id", "local_raw_path", "quality_status", "review_notes"]].head(20))
else:
    print("No review-needed candidates available yet.")
```

## Interpretation

Hard rejects are not deleted; they are recorded for traceability. Review-needed candidates require human judgment before any use in anomaly training.
