# Query Catalog Review

Purpose: review the editable query catalog for likely-normal tyre web candidates.

Inputs: `configs/web_collection/query_catalog.yaml`.

Outputs: summary tables for query priority, view types, and negative keyword coverage.

Current status: scaffold is implemented; no provider search has been run.

```python
from pathlib import Path
import pandas as pd
import yaml

CATALOG = Path("configs/web_collection/query_catalog.yaml")
assert CATALOG.exists(), "Run scripts/web_collection/generate_query_catalog.py first."
payload = yaml.safe_load(CATALOG.read_text())
queries = pd.DataFrame(payload["queries"])
queries.head()
```

```python
print("Total queries:", len(queries))
display(queries.groupby(["priority", "view_type"]).size().reset_index(name="count"))
```

```python
display(queries[["query_id", "query_text", "priority", "view_type", "notes"]].sort_values(["priority", "query_id"]))
```
