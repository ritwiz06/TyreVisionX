# Human Review Workflow Demo

Purpose: demonstrate the review status workflow for web-image candidates.

Inputs: `reports/web_collection/review_queue_template.csv` or a generated review queue.

Outputs: status counts and a reviewer-facing table.

Current status: template/demo only; no real web candidates have been reviewed.

```python
from pathlib import Path
import pandas as pd

QUEUE = Path("data/interim/web_curated_candidates/review_queue.csv")
TEMPLATE = Path("reports/web_collection/review_queue_template.csv")
path = QUEUE if QUEUE.exists() else TEMPLATE
df = pd.read_csv(path)
print("Loaded", path)
display(df.head())
```

```python
status_order = [
    "pending_review",
    "approved_likely_normal",
    "rejected_irrelevant",
    "rejected_low_quality",
    "rejected_likely_defect",
    "uncertain_holdout",
]
counts = df["human_review_status"].value_counts().reindex(status_order, fill_value=0)
display(counts.to_frame("count"))
counts.plot(kind="bar", title="Human review statuses")
```

## Reviewer Rule

Only `approved_likely_normal` candidates should be considered for future good-only anomaly training manifests. `uncertain_holdout` can be useful later for stress testing, but it should not be mixed into normal training.
